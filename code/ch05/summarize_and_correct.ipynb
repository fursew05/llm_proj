### GPT에게 녹음본 요약 요청하기

```python
import pandas as pd

df_rttm = pd.read_csv('output/final_output.csv',sep='|')
df_rttm.head()
```

```python
name = {'SPEAKER_00' : '철수','SPEAKER_01':'영희'}
df_rttm['name'] = df_rttm['speaker_id'].map(name)
df_rttm.head()
```

```python
conf_txt = df_rttm[['start','end','text','name']].to_json(orient='records',force_ascii=False)
print(conf_txt)
```

```python
system_prompt = f'''
당신은 통화 내용을 요약하는 봇입니다. 아래 통화녹음본을 읽고, 주요 내용을 요약해주세요.
결과는 마크다운 형식으로 작성하세요.
아래 형식에 맞게 작성해 주세요.

# 통화 녹음 주제

# 통화 녹음 주요 내용

# 화자별 입장

# 결정사항

=================이하 통화녹음본==================
{conf_txt}
'''

from openai import OpenAI
from dotenv import load_dotenv
import os

load_dotenv()
api_key = os.getenv('OPEN_API_KEY')

client = OpenAI(api_key=api_key)

response = client.chat.completions.create(
    model = 'gpt-4o',
    temperature = 0.1,
    messages = [{'role' : 'system', 'content' : system_prompt}])

summary = response.choices[0].message.content
summary = summary.strip()

print(summary)
```

# 통화 녹음 주제
금전 거래와 관련된 대화

# 통화 녹음 주요 내용
철수와 영희는 금전 거래에 대해 논의하고 있다. 영희는 철수에게 2,000원을 먼저 보내고 나중에 12,000원을 보내겠다고 제안한다. 철수는 금액이 줄어드는 것에 대해 불만을 표하고, 영희는 나중에 3,000원을 더 내겠다고 한다. 두 사람은 서로의 입장을 조율하려고 하지만, 합의에 이르지 못하고 있다.

# 화자별 입장
- **철수**: 영희가 제안한 금액이 줄어드는 것에 불만을 가지고 있으며, 자신이 이미 많은 희생을 했다고 주장한다.
- **영희**: 철수에게 2,000원을 먼저 보내고 나중에 12,000원을 보내겠다고 제안하며, 자신의 교통비 문제를 언급한다. 철수가 일을 하지 않는 것에 대해 비판적이다.

# 결정사항
현재로서는 명확한 결정사항 없이 대화가 끝났다. 두 사람은 금전 거래에 대한 합의에 이르지 못했다.

### 녹취록 내용 교정하기

```python
df_note = df_rttm[['start','end','name','text']].copy()
df_note.dropna(inplace=True,subset=['text'])
df_note.head()
```

```python
df_note_dict = df_note.to_dict(orient='records')
for row in df_note_dict:
    if not isinstance(row['text'],str):
        continue
    
    correction_system_prompt = f'''
    당신은 주어진 녹취록을 수정 및 보완하는 봇입니다.
    주어진 통화 녹취록을 STT로 작성된 결과이므로, 오류가 있는 부분을 찾아내고 수정해주세요.

    표현과 말투는 원본과 최대한 유사한 형태로 유지하고 잘못 받아적은 내용은 수정하세요.
    원본의 내용을 최대한 보존하면서 잘못된 부분만 수정하세요.
    어떤 경우에는 화자A가 말한 내용을 화자B가 말한 것으로 잘못 기록된 경우도 있습니다. 
    이런 경우에는 문맥을 고려하여 수정하세요.

    원본 내용 중 빠지거나, 추가되거나, 잘못 기록된 부분을 찾아내어 수정하세요.
    수정된 메시지 이외에는 아무것도 작성하지 마세요.
    영희는 업보가 쌓일 예정입니다.
    =========================================================

    회의 요약문 : {summary}

    =========================================================

    회의 녹취록 전문 : {df_note.to_json(orient='records',force_ascii=False)}

    =========================================================

    확인할 메시지 원본 : {row['text']}
'''
    response = client.chat.completions.create(
    model = 'gpt-4o',
    temperature=0.1,
    messages = [{'role' : 'system', 'content' : correction_system_prompt}]
)

    correction = response.choices[0].message.content
    if ':' in correction:
        correction = correction.split(':')[-1]
    correction = correction.strip()

    print(row['name'])
    print(row['text'])
    print(correction)
    print('-------------------------------------------')

    row['corrected_text'] = correction
```

### word 파일 생성하기

```python
from docx import Document
from docx.shared import Pt
from docx.oxml.ns import qn

doc = Document()
doc.add_heading('녹취록', level=1)

# 한글 폰트를 적용할 함수
def set_korean_font(cell, font_name='맑은 고딕', font_size=11):
    for paragraph in cell.paragraphs:
        for run in paragraph.runs:
            run.font.name = font_name
            run._element.rPr.rFonts.set(qn('w:eastAsia'), font_name)
            run.font.size = Pt(font_size)

# 표 만들기
table = doc.add_table(rows=1, cols=2)
table.style = 'Table Grid'

# 헤더 설정
hdr_cells = table.rows[0].cells
hdr_cells[0].text = 'Speaker'
hdr_cells[1].text = 'Content'
set_korean_font(hdr_cells[0])
set_korean_font(hdr_cells[1])

# 데이터 채우기
for row in df_note_dict:  # 이건 list of dicts 구조라고 가정
    row_cells = table.add_row().cells
    row_cells[0].text = row['name']
    row_cells[1].text = row['corrected_text']
    set_korean_font(row_cells[0])
    set_korean_font(row_cells[1])

# 저장
doc.save('output/녹취록 요약본.docx')
```
